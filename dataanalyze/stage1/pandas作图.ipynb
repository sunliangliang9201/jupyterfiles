```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
##pandas
#
s = Series(np.random.randn(10).cumsum(),index=np.arange(0,100,10))
s.plot()

df = DataFrame(np.random.randn(10, 4).cumsum(0),
              columns=['A', 'B', 'C', 'D'],
              index=np.arange(0,100,10))
df.plot()

#
fig, axes = plt.subplots(2,1)
data = Series(np.random.rand(16),index=list('abcdefghijklmnop'))
data.plot(kind='bar',ax=axes[0],color='k',alpha=0.7)
data.plot(kind='barh',ax=axes[1],color='r',alpha=0.7)

df = DataFrame(np.random.rand(6,4),index=['one', 'two', 'three', 'four', 'five', 'six'],
              columns=pd.Index(['A', 'B', 'C', 'D'],name = 'Genus'))
df
df.plot(kind='bar')

plt.figure()
df.plot(kind='barh',stacked=True,alpha=0.5)

tips = pd.read_csv(r'e:\data\tips.csv')
party_counts = pd.crosstab(tips.day, tips['size'])
party_counts
party_counts.plot(kind='bar', stacked=True)

#
plt.figure()
tips['tip_pct'] = tips['tip']/tips['total_bill']
tips['tip_pct'].hist(bins=50)
plt.figure()
tips['tip_pct'].plot(kind='kde')

#
plt.figure()
macro = pd.read_csv('e:\data\macrodata.csv')
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data = np.log(data).diff().dropna()
trans_data[-5:]
plt.scatter(trans_data['m1'], trans_data['unemp'])
plt.title('Changes in log %s vs. log %s' % ('m1','unemp'))
pd.scatter_matrix(trans_data, diagonal='kde', color='k', alpha=0.3)
```
