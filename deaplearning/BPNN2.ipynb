```python
#BPNN：手写体识别，单个数字
import numpy as np
from sklearn.datasets import load_digits
from sklearn.preprocessing import LabelBinarizer#标签二值化
from sklearn.cross_validation import train_test_split#切分数据为训练数据和测试数据
import pylab as pl

# #载入数据集
# digits = load_digits()
# print(digits.data.shape)
# #看一个手写体图片
# pl.gray()
# pl.matshow(digits.images[0])
# pl.show()

def sigmoid(x):
    return 1/(1+np.exp(-x))

def dsigmoid(x):
    return x*(1-x)

class NetualNetwork(object):
    def __init__(self,layers):#(64,100,10):
        #权值初始化
        self.V = np.random.random((layers[0]+1,layers[1]+1))*2-1
        self.W = np.random.random((layers[1]+1,layers[2]))*2-1
    def train(self,X,Y,lr=0.11,epochs=20000):
        #添加偏置值
        temp = np.ones([X.shape[0],X.shape[1]+1])
        temp[:,0:-1] = X
        X = temp

        for n in range(epochs+1):
            #当有很多数据需要训练的时候，一般是分组批次来训练，这里就不分组了
            i = np.random.randint(X.shape[0])#随机选取一个数据
            x = [X[i]]
            x = np.atleast_2d(x)#转换为二维数据

            L1 = sigmoid(np.dot(x,self.V))
            L2 = sigmoid(np.dot(L1,self.W))

            L2_delta = (Y[i]-L2)*dsigmoid(L2)
            L1_delta = L2_delta.dot(self.W.T)*dsigmoid(L1)

            self.W += lr*L1.T.dot(L2_delta)
            self.V += lr*x.T.dot(L1_delta)
            #每训练1000次预测一次准确率
            if n%1000 == 0:
                predictions = []
                for j in range(X_test.shape[0]):
                    o = self.predict(X_test[j])
                    predictions.append(np.argmax(o))#获取预测结果
                accuracy = np.mean(np.equal(predictions,y_test))
                print('epoch:',n,'accuracy:',accuracy)
    def predict(self,x):
        temp = np.ones(x.shape[0]+1)
        temp[0:-1] = x
        x = temp
        x = np.atleast_2d(x)

        L1 = sigmoid(np.dot(x, self.V))
        L2 = sigmoid(np.dot(L1, self.W))
        return L2

digits = load_digits()
X = digits.data#把数据存入X
Y = digits.target#标签
#输入数据归一化
X -= X.min()
X /=(X.max()-X.min())

nn = NetualNetwork([64,100,10])#创建网络
X_train,X_test,y_train,y_test = train_test_split(X,Y)#分割数据1/4为测试数据，3/4为训练数据
#标签二值化,因为原始标签是与输入一致的，比如输入5，标签也是5，但是我们一般是这样，比如输入5，那么除了第五个输出神经元输出为1，其他为0，也就是0000100000
labels_train = LabelBinarizer().fit_transform(y_train)
labels_test = LabelBinarizer().fit_transform(y_test)

print('start')
nn.train(X_train,labels_train,epochs=20000)
print('end')


```
